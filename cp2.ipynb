## 1. $n = 50$, $x = [1, ..., 1]$, so that $b$ is the sum of each column/row of $H$

```r
n = 50
H = matrix(0,n,n)
for (i in 1:n){
    for (j in 1:n){
        H[i,j] = 1/(i+j-1)
    }
}

x = matrix(1,n,1)

b = H%*%x

H[1:5,1:5]
x[1:5,]
b[1:5,]

trh = 0
for (i in 1:n){
    trh = trh + as.matrix(H%*%H)[i,i]
}
a0 = H/trh
```

Set $k = 100$

```r
k = 100
est_x = matrix(0,n,k)
for (i in 1:k){
    a0 = a0%*%as.matrix(2*diag(n)-as.matrix(H%*%a0))
    est_x[,i] = matrix(a0%*%b,n,1)
}
plot(1:k,est_x[1,]-1,ylim = c(-.5,.5),cex = 0, xlab = 'Number of iteration', ylab = 'Residual of estimated x Value', main = 'Process of converging when n = 50')
for (i in 1:n){
    lines(1:k,est_x[i,]-1, col = i)
}
```

When $n=50, x= [1, ..., 1]$, from the plot we could see that $x_k$ converges to $x$ before about 50 iterations, but diverges after then.

#  4. $n = 50$, $b = [1, ..., 0]$

```r
n = 50
H = matrix(0,n,n)
for (i in 1:n){
    for (j in 1:n){
        H[i,j] = 1/(i+j-1)
    }
}

b = matrix(0,n,1)
b[1,1] = 1

H[1:5,1:5]
b[1:5,]

trh = 0
for (i in 1:n){
    trh = trh + as.matrix(H%*%H)[i,i]
}
a0 = H/trh
```

```r
k = 100
est_x = matrix(0,n,k)
for (i in 1:k){
    a0 = a0%*%as.matrix(2*diag(n)-as.matrix(H%*%a0))
    est_x[,i] = matrix(a0%*%b,n,1)
}
plot(1:k,est_x[1,],ylim = c(-100,100),cex = 0, xlab = 'Number of iteration', ylab = 'Estimated x Value', main = 'Process of converging when n = 50')
for (i in 1:n){
    lines(1:k,est_x[i,]-1, col = i)
}
```

When $n = 50$, $b = [1, ..., 0]$, the result doesn't converge.
